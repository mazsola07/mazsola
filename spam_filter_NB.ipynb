```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from itertools import chain
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None)
df.columns = ['Label', 'SMS']
df
```

```python
plt.figure(figsize=(12,8))
# sns.countplot() - show the counts of observations in each categorical bins using bars
sns.countplot(x = df.Label)
plt.yticks([])
# Series.value_counts() - return a Series containing counts of unique values
df.Label.value_counts(normalize=True)
```

```python
# DataFrame.sample(n,frac) - return a new object of same type as caller containing n items 
#                            randomly sampled from the caller object 
#                            (return a random sample of items from an axis of object)
#    n - number of items from axis to return
#    frac - fraction of axis items to return (cannot be used with n)
randomized = df.sample(frac=1, random_state=1)
train_size = round(len(randomized)*0.8)
# 'drop=False' - insert a column 'Index' with values of old indeces
train = randomized[:train_size].reset_index(drop=True)
test = randomized[train_size:].reset_index(drop=True)
train.shape, test.shape
```

```python
# Series.str.replace(pat,repl,regex) - replace each occurence of pattern/regex in Series
#   pat - string can be a character sequence or regular expression
#   repl - replacement string
# re - regular expression operations
# \W - matches any character which is not a word character
train.SMS = train['SMS'].str.replace('\W', ' ', regex=True)
train.SMS = train['SMS'].str.lower()
train.SMS.head(10)
```

```python
# Series.str.split(pat) - split strings in the Series
#    pat - string or reg exp to split on; if not specified split on whitespace
messages = train['SMS'].str.split()
messages
```

```python
# itertools - functions creating iterators for efficient looping
# itertools.chain(*iterables) - make an iterator that returns elements from the first iterable 
#                               until it is exhausted, then proceeds to the next iterable, until 
#                               all of the iterables are exhausted
#                               (Used for treating consecutive sequences as a single sequence)
words = list(chain(*messages))
# Series.unique() - return unique values of Series object as a numpy array
vocabulary = pd.Series(words).unique()
len(vocabulary)
```

```python
no_of_messages = len(train.SMS)
word_counts = {word: [0]*no_of_messages for word in vocabulary}
for index, mssge in enumerate(messages):
    for word in mssge:
        word_counts[word][index] += 1
```

Document-term matrix - a mapping of every word to its message and the number of times the word                        occurs in the message
                       rows: words in the dictionary
                       columns: messages

```python
# nltk - natural language toolkit
# nltk.tokenize.RegexpTokenizer() - splits a string into substrings using a regular expression
#   + - causes the resulting regexp to match 1 or more repetitions of the preceding regexp
token = RegexpTokenizer(r'[a-z0-9A-Z]+')
# CountVectorizer() - convert a collection of text documents to a matrix of token counts
#    lowercase=True - convert all characters to lowercase before tokenizing
#    stop_words='english' - a built-in stop word list for English is used 
#    ngram_range=(1,1) - only unigrams (single words or terms)
vectorizer = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1,1), 
                            tokenizer=token.tokenize)
# fit_transform() - learn the vocabulary dictionary and return document-term matrix
counts = vectorizer.fit_transform(train.SMS)
# CountVectorizer.get_feature_names_out() - get output feature names for transformation
print(f'length of vocabulary after removing stopwords {len(vectorizer.get_feature_names_out())}')
vocabulary = vectorizer.get_feature_names_out()
word_counts = dict(zip(vocabulary, np.transpose(counts.toarray())))
```

```python
vector_space = pd.DataFrame(word_counts)
train = pd.concat([train, vector_space], axis=1)
train.head(3)
```

```python
alpha = 1 # Laplace smoothing constant 
P_spam = train.Label.value_counts(normalize=True)['spam']
P_ham = train.Label.value_counts(normalize=True)['ham']
# number of words in spam messages
N_spam = len(list(chain(*messages[train.Label == 'spam'])))
# number of words in ham messages
N_ham = len(list(chain(*messages[train.Label == 'ham'])))
N_vocab = len(vocabulary)
```

```python
P_word_given_spam = {word: 0 for word in vocabulary}
P_word_given_ham = {word: 0 for word in vocabulary}

Spam_messages = train[train.Label == 'spam']
Ham_messages = train[train.Label == 'ham']

for word in vocabulary:
    N_word_given_spam = Spam_messages[word].sum()
    N_word_given_ham = Ham_messages[word].sum()
    P_word_given_spam[word] = (
        (N_word_given_spam + alpha) / (N_spam + (alpha * N_vocab))
    )
    P_word_given_ham[word] = (
        (N_word_given_ham + alpha) / (N_ham + (alpha * N_vocab))
    )
```

```python
def classify(message,verbose):
    # re.sub(pattern, repl, string) - return the string obtained by replacing the leftmost 
    #                                 non-overlapping occurences of pattern in string by the
    #                                 replacement repl
    message = re.sub('\W',' ',message)
    message = message.lower()
    message = message.split()
    spam = 1
    ham = 1
    for word in message:
        if word in P_word_given_spam.keys():
            spam *= P_word_given_spam[word]
        if word in P_word_given_ham.keys():
            ham *= P_word_given_ham[word]
        
    P_spam_given_message = P_spam * spam
    P_ham_given_message = P_ham * ham
    
    if verbose:
        print("P(spam|message) = ",P_spam_given_message)
        print("P(ham|message) = ",P_ham_given_message)
    
    if P_spam_given_message > P_ham_given_message:
        if verbose:
            print('Label: spam')
        return 'spam'
    elif P_spam_given_message < P_ham_given_message:
        if verbose:
            print('Label: ham')    
        return 'ham'
    else:
        if verbose:
            print('Human assistance needed, equal probabilities')
        return 'Not classified'
```

```python
# Series.apply(func) - invoke functions on values of Series
test['predicted'] = test.SMS.apply(classify, verbose=0)
test.head(5)
```
